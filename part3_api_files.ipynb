**Part B — Read:**

```python
print("\n--- Reading File Contents ---")
with open("python_notes.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    for index, line in enumerate(lines, 1):
        print(f"{index}. {line.strip()}")

print(f"\nTotal number of lines: {len(lines)}")

# Keyword Search
keyword = input("\nEnter a keyword to search for: ").lower()
found = False
for line in lines:
    if keyword in line.lower():
        print(f"Match found: {line.strip()}")
        found = True

if not found:
    print(f"No lines found containing '{keyword}'.")
```

**Task 2 — API Integration**

```python
import requests
```

**Step 1 — Fetch and Display Products:**

```python
url = "https://dummyjson.com/products?limit=20"
response = requests.get(url)
data = response.json()
products = data['products']

print(f"{'ID':<4} | {'Title':<30} | {'Category':<15} | {'Price':<8} | {'Rating':<6}")
print("-" * 75)
for p in products:
    print(f"{p['id']:<4} | {p['title'][:30]:<30} | {p['category']:<15} | ${p['price']:<7} | {p['rating']:<6}")
```

**Step 2 — Filter and Sort:**

```python
# filtering the product based on rating >= 4.5
filtered_products = [p for p in products if p['rating'] >= 4.5]
#Sorting the filtered products based on price in descending order
sorted_products = sorted(filtered_products, key=lambda x: x['price'], reverse=True)

print("\n--- Filtered (Rating >= 4.5) & Sorted by Price (Desc) ---")
for p in sorted_products:
    print(f"${p['price']}: {p['title']} ({p['rating']})")
```

**Step 3 — Search by Category:**

```python
print("\n--- Laptops Category ---")
laptop_res = requests.get("https://dummyjson.com/products/category/laptops")
laptops = laptop_res.json()['products']
for laptop in laptops:
    print(f"Name: {laptop['title']} | Price: ${laptop['price']}")
```

**Task 3 — Exception Handling**

**Part A — Guarded Calculator:**

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print("\n--- Calculator Tests ---")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))
```

**Part B — Guarded File Reader:**

```python
def read_file_safe(filename):
    try:
        with open(filename, "r") as f:
            return f.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")

print("\n--- File Reader Tests ---")
read_file_safe("python_notes.txt")
read_file_safe("ghost_file.txt")
```

**Part C & D: Robust API & Validation Loop**

```python
def fetch_product_interactive():
    while True:
        user_input = input("\nEnter a product ID to look up (1–100), or 'quit' to exit: ").strip().lower()
        if user_input == 'quit':
            break

        if not user_input.isdigit() or not (1 <= int(user_input) <= 100):
            print("Warning: Please enter a valid integer between 1 and 100.")
            continue

        product_id = int(user_input)
        try:
            res = requests.get(f"https://dummyjson.com/products/{product_id}", timeout=5)
            if res.status_code == 404:
                print("Product not found.")
            elif res.status_code == 200:
                p_data = res.json()
                print(f"Title: {p_data['title']} | Price: ${p_data['price']}")
        except requests.exceptions.ConnectionError:
            print("Connection failed. Please check your internet.")
        except requests.exceptions.Timeout:
            print("Request timed out. Try again later.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

fetch_product_interactive()
```

**Task 4 — Logging to File**

```python
import os
from datetime import datetime

def log_error(function_name, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_entry = f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n"
    with open("error_log.txt", "a") as f:
        f.write(log_entry)

# Trigger 1: ConnectionError
try:
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=2)
except Exception as e:
    log_error("fetch_products", type(e).__name__, str(e))

# Trigger 2: HTTP 404
res = requests.get("https://dummyjson.com/products/999")
if res.status_code != 200:
    log_error("lookup_product", "HTTPError", f"{res.status_code} Not Found for product ID 999")

print("\n--- Contents of error_log.txt ---")
if os.path.exists("error_log.txt"):
    with open("error_log.txt", "r") as f:
        print(f.read())
```
